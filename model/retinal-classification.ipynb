```python
# augument training data
train_datagen = ImageDataGenerator(rescale = 1./255.,
                                    rotation_range = 30,
                                    width_shift_range = 0.1,
                                    height_shift_range = 0.1,
                                    zoom_range = 0.1,
                                    horizontal_flip = True
                                   )
train_generator = train_datagen.flow_from_directory('./Training', 
                                                    target_size=(150,150), 
                                                    class_mode='binary')
```

```python
# augument validation set
validation_datagen = ImageDataGenerator(rescale=1/255)
validation_generator = validation_datagen.flow_from_directory('./Validation', 
                                                              target_size=(150,150),
                                                              class_mode='binary')
```

```python
# augument test set
test_datagen = ImageDataGenerator(rescale=1/255)
test_generator = test_datagen.flow_from_directory('./Testing', 
                                                              target_size=(150,150),
                                                              class_mode='binary')
```

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten,Dense,Dropout
from tensorflow.keras.optimizers import RMSprop
from tensorflow import keras
```

```python
# load pretrained model
model = EfficientNetB0(input_shape=(150,150,3), weights='../input/efficientnet/efficientnetb0_notop.h5', include_top=False)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
x = model.output
x = Flatten()(x)
x = Dropout(0.5)(x)
x = Dense(1024, activation="relu")(x)
predictions = Dense(1, activation="sigmoid")(x)
```

```python
m = Model(inputs=model.input, outputs=predictions)
m.summary()
```

```python
m.compile(optimizer = RMSprop(learning_rate=0.0001), 
              loss = 'binary_crossentropy', 
              metrics = ['accuracy'])
```

```python
results = m.fit(train_generator,epochs=20,
                              steps_per_epoch=20,
                              validation_data=validation_generator,
                             validation_steps=5)
```

```python
m.evaluate(test_generator)
```
